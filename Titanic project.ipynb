```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\azhar\OneDrive\Attachments\Desktop\Titanic.csv")
print(df.head())
```

```python
df.columns = df.columns.str.strip() 
```

```python
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived')
plt.show()
```

```python
Observation: Survivors tended to be younger and had higher fare tickets.
```

```python
print(df.columns)
df.columns = df.columns.str.strip().str.lower()
```

```python
sns.heatmap(df[['age', 'fare', 'pclass', 'survived']].corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
Observation: Fare and Pclass are negatively correlated. Fare has a mild positive correlation with survival.
```

```python
df['age'].hist(bins=20)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
Observation: Most passengers were between 20–40 years old.
```

```python
sns.boxplot(x='survived', y='age', data=df)
plt.title('Age Distribution by Survival')
plt.show()
```

```python
Observation: Survivors had a lower average age than non-survivors.
```

```python
sns.scatterplot(x='age', y='fare', hue='survived', data=df)
plt.title('Fare vs Age colored by Survival')
plt.show()
```

```python
Observation: Higher fare and younger passengers had a better survival rate.
```
